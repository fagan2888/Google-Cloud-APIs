## Operations using `google.cloud.storage`

```python
from google.cloud import storage
client = storage.Client()

bucket = client.get_bucket('deep-nexus')
```

Listing objects:

In the first example below I am using delimiter to list objects directly under the directory specified rather than all of the objects under that tree structure. 

```python
blobs = bucket.list_blobs(prefix='Kather_texture/raw-data/', delimiter='/')
for blob in blobs:
  print(blob.name)
```

If we didn't use delimiter then everything would be listed:

```python
blobs = bucket.list_blobs(prefix='Kather_texture/raw-data/')
for blob in blobs:
  print(blob.name)
```

### Downloading the files to Datalab
Downloding a single file

```python
blob = bucket.blob('Kather_texture/raw-data/Kather_texture_2016_larger_images_10/CRC-Prim-HE-09_APPLICATION.tif')

blob.download_to_filename('CRC-Prim-HE-09_APPLICATION.tif')
```
