### Fetch image paths, add labels, and export paths with labels to a .csv file

```python
!gsutil -m cp -r gs://deep-nexus/dogs_and_cats/train_1000 gs://deep-nexus-vcm/train_1000 
```

```python
cat = !gsutil ls gs://deep-nexus-vcm/train_1000
```

```python
len(cat)
```

```python
import pandas as pd
```

```python
cat_1000 = pd.DataFrame(cat, columns=['link'])
cat_1000['label'] = 'cat'
cat_1000.head()
```

```python
cat_1000.to_csv('cat_1000_name.csv', header=False, index=False)
```

```python
dognames
```
